# 7 - ILS

ILS stands for "Iterated Local Search" and it's a conformational search algorithm. In short, this algorithm is divided in two main components, the inner and outer loop. The inner loop can be, for example, a Monte Carlo simulation, and should explore and sample the local minimum, while the outer loop performs a "jump": a large conformational change, taking the system to a whole different local minimum for the next inner loop iteration. In this way, the conformational space can be efficiantly sampled, with a higher certainty of exploring the global minimum during the simulation. In ProtoSyn, this algorithm is employed as a Driver, and will be briefly explored in this example. As a starting point, we will load the 2A3D structure, copy its sequence and build a completly stretched peptide.

```python
using ProtoSyn
using ProtoSyn.Units
```

```python
base_truth = ProtoSyn.Peptides.load("data/2a3d.pdb")
```

```python
sequence = ProtoSyn.Peptides.sequence(base_truth)
```

```python
residue_library = ProtoSyn.Peptides.grammar;
pose = ProtoSyn.Peptides.build(residue_library, [string(aminoacid) for aminoacid in sequence])
```

```python
using Bio3DView

ProtoSyn.write(pose, "output/monte_carlo.pdb")
style = Style("stick")
viewfile("output/monte_carlo.pdb", style = style)
```

Note that Prolines naturally introduce a "kink" in the structure. In order to limit the degrees of freedom of our simulation, we can consider the secondary structures as immutable, sampling only the loop regions for the correct fold. For this, we can firstly apply a secondary structure to our candidate structure in the helical regions. In this example, we will simply copy the "blocks" of secondary structure from the 2A3D structure. If the base truth structure was unkown, several secondary structure prediction machine learning models have been proposed in past, and released under the form of prediction servers (RaptorX, for example).

```python
ProtoSyn.Peptides.setss!(pose, ProtoSyn.Peptides.SecondaryStructure[:helix], rid"2:19")
ProtoSyn.Peptides.setss!(pose, ProtoSyn.Peptides.SecondaryStructure[:helix], rid"27:44")
ProtoSyn.Peptides.setss!(pose, ProtoSyn.Peptides.SecondaryStructure[:helix], rid"49:71")
```

```python
ProtoSyn.write(pose, "output/monte_carlo.pdb")
viewfile("output/monte_carlo.pdb", style = style)
```

As a final preparation step, and for a first approach to a folding problem, one extra step can be taken to reduce the degrees of freedom of the system: to remove the sidechains. This greatly reduces clashes, at the expense of higher resolutions. The sidechains can latter be re-added and packaged, in a refinement stage.

```python
ProtoSyn.Peptides.remove_sidechains!(pose, residue_library)
```

```python
ProtoSyn.write(pose, "output/monte_carlo.pdb")
viewfile("output/monte_carlo.pdb", style = style)
```

The objetive of this example is, therefore, to use the ILS algorithm to explore the conformational space of the remaining loop regions and place these "blocks" in the 3D space, folding the peptide. For this, we will need to define the inner loop (as a Monte Carlo Driver), the evaluator Energy Function and a "jump" Mutator, responsible to perform a "large enough" conformational change.

```python
energy_function = ProtoSyn.Common.default_energy_function()
```

```python
selection            = (an"C" | an"N") & (rid"20:26" | rid"45:48")
probability_mutation = 1/count(selection(pose))
inner_driver_mutator = ProtoSyn.Mutators.DihedralMutator(randn, probability_mutation, 1.0, selection)
```

Note that only the loop regions are being selected for conformational changes.

```python
inner_driver_callback = ProtoSyn.Common.default_energy_step_callback(100)
```

```python
inner_driver_thermostat = ProtoSyn.Drivers.get_linear_quench(0.8, 2000)
```

```python
inner_driver = ProtoSyn.Drivers.MonteCarlo(energy_function, inner_driver_mutator, inner_driver_callback, 2000, inner_driver_thermostat)
```

```python
jump = ProtoSyn.Mutators.DihedralMutator(randn, 1.0, 0.2, selection)
```

```python
outer_driver_callback = ProtoSyn.Common.default_energy_step_callback(1)
```

```python
outer_driver_thermostat = ProtoSyn.Drivers.get_constant_temperature(0.4)
```

```python
ils = ProtoSyn.Drivers.ILS(energy_function, jump, inner_driver, outer_driver_callback, 10, outer_driver_thermostat)
```

All that's left to do is launch the simulation and analyze the results.

```python
ils(pose)
```

```python
ProtoSyn.write(pose, "output/monte_carlo.pdb")
viewfile("output/monte_carlo.pdb", style = style)
```

We can now align the resulting pose to the base truth and measure the RMSD value. Note that, for a quick low resolution simulation such as this example, it's expected to obtain RMSD values between 6 and 10 angstrom. Further refinement stages can be performed from this resulting structure.

```python
ProtoSyn.align!(pose, base_truth, an"CA")
```

```python
ProtoSyn.rmsd(pose, base_truth, an"CA")
```

## Conclusion

In this example we were able to perform a low resolution folding prediction of a small peptide (under 10 Angstrom RMSD) using the "Iterated Local Search" algorithm (ILS). In a real application, the algorithm should run for a greatly extended number of steps and inner loop steps, as well as include more specialized Mutators as samplers.
